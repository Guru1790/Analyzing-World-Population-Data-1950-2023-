```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
df = pd.read_csv('/content/World Population.csv')
```

```python
# Renaming columns for convenience
df.columns = ['Rank', 'Country', 'Capital', 'Continent', 'Area_km2', '1950', '1955', '1960', '1965', '1970', '1975', '1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015', '2020', '2023', 'World_Population_Percentage']
```

```python
# Handling missing values (if any)
df = df.dropna()
```

```python
# Convert population columns to numeric
for year in range(1950, 2024, 5):
    df[str(year)] = pd.to_numeric(df[str(year)], errors='coerce')
```

Data Analysis

```python
# Summary Statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Top 10 countries by 2023 population
top_10_2023 = df.nlargest(10, '2023')
print("\nTop 10 countries by 2023 population:")
print(top_10_2023[['Country', '2023']])
```

```python
# Bottom 10 countries by 2023 population
bottom_10_2023 = df.nsmallest(10, '2023')
print("\nBottom 10 countries by 2023 population:")
print(bottom_10_2023[['Country', '2023']])
```

```python
# Population Growth Analysis
df['Growth_1950_2023'] = ((df['2023'] - df['1950']) / df['1950']) * 100
top_growth_countries = df.nlargest(10, 'Growth_1950_2023')
print("\nTop 10 countries by population growth (1950-2023):")
print(top_growth_countries[['Country', 'Growth_1950_2023']])
```

```python
# Population Growth Rate Analysis
for year in range(1955, 2024, 5):
    df[f'Growth_{year-5}_{year}'] = ((df[str(year)] - df[str(year-5)]) / df[str(year-5)]) * 100

print("\nPopulation Growth Rate Analysis:")
print(df.head())
```

```python
# Continent-Wise Analysis
continent_population_2023 = df.groupby('Continent')['2023'].sum().sort_values(ascending=False)
print("\nContinent-Wise Population in 2023:")
print(continent_population_2023)
```

```python
# Correlation Analysis
population_columns = [str(year) for year in range(1950, 2024, 5)]
correlation_matrix = df[population_columns].corr()
print("\nCorrelation Matrix of Population Over Years:")
print(correlation_matrix)
```

```python
# Descriptive Statistics for Each Continent
continent_stats = df.groupby('Continent')[population_columns].describe()
print("\nDescriptive Statistics for Each Continent:")
print(continent_stats)
```

```python
# Checking for any outliers in the dataset using IQR
Q1 = df[population_columns].quantile(0.25)
Q3 = df[population_columns].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df[population_columns] < (Q1 - 1.5 * IQR)) | (df[population_columns] > (Q3 + 1.5 * IQR))).sum().sum()
print("\nNumber of Outliers in Population Data:")
print(outliers)
```

```python
# Check for any countries with decreasing population over the years
decreasing_population_countries = df[df[population_columns].diff(axis=1).dropna(axis=1).apply(lambda x: (x < 0).all(), axis=1)]
print("\nCountries with Decreasing Population Over the Years:")
print(decreasing_population_countries[['Country'] + population_columns])
```

Data Visualization

```python
# Population Growth Over Time
plt.figure(figsize=(14, 8))
for country in top_10_2023['Country']:
    plt.plot(df.columns[5:-1], df[df['Country'] == country].iloc[0, 5:-1], label=country)

plt.title('Population Growth Over Time (Top 10 countries by 2023 population)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
# Population Distribution by Continent
continent_population_2023 = df.groupby('Continent')['2023'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=continent_population_2023.values, y=continent_population_2023.index)
plt.title('Population Distribution by Continent (2023)')
plt.xlabel('Population')
plt.ylabel('Continent')
plt.show()
```

```python
# World Population Percentage
plt.figure(figsize=(14, 8))
sns.barplot(x='World_Population_Percentage', y='Country', data=top_10_2023)
plt.title('World Population Percentage (Top 10 countries by 2023 population)')
plt.xlabel('World Population Percentage')
plt.ylabel('Country')
plt.show()
```

```python
# Heatmap for Correlation Matrix
plt.figure(figsize=(14, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix of Population Over Years')
plt.show()
```

```python
# Population Growth Rate Over Time for Top Growth Countries
plt.figure(figsize=(14, 8))
for country in top_growth_countries['Country']:
    growth_rate_columns = [f'Growth_{year-5}_{year}' for year in range(1955, 2024, 5)]
    plt.plot(growth_rate_columns, df[df['Country'] == country].iloc[0][growth_rate_columns], label=country)

plt.title('Population Growth Rate Over Time (Top 10 countries by population growth)')
plt.xlabel('Growth Rate Period')
plt.ylabel('Growth Rate (%)')
plt.legend()
plt.show()
```

```python
# Boxplot for Population Distribution by Continent
plt.figure(figsize=(14, 8))
sns.boxplot(x='Continent', y='2023', data=df)
plt.title('Population Distribution by Continent (2023)')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.show()
```

```python
# Histogram for Population Distribution in 2023
plt.figure(figsize=(14, 8))
plt.hist(df['2023'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Population Distribution in 2023')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line Plot for Population Trends of Selected Countries
selected_countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan']
plt.figure(figsize=(14, 8))
for country in selected_countries:
    plt.plot(df.columns[5:-1], df[df['Country'] == country].iloc[0, 5:-1], label=country)

plt.title('Population Trends of Selected Countries (1950-2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
# Top 10 Countries by Population for Each Decade
years = [str(year) for year in range(1950, 2024, 5)]
for year in years:
    top_10 = df.nlargest(10, year)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=year, y='Country', data=top_10)
    plt.title(f'Top 10 Countries by Population in {year}')
    plt.xlabel('Population')
    plt.ylabel('Country')
    plt.show()
```

```python
# Population Distribution Over Time for Each Continent
continents = df['Continent'].unique()
plt.figure(figsize=(14, 8))
for continent in continents:
    continent_data = df[df['Continent'] == continent]
    plt.plot(years, continent_data[years].sum(), label=continent)

plt.title('Population Distribution Over Time for Each Continent')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
# Pairplot for Selected Years
selected_years = ['1950', '2000', '2023']
sns.pairplot(df[selected_years])
plt.suptitle('Pairplot for Selected Years', y=1.02)
plt.show()
```

```python
# Violin Plot for Population Distribution by Continent
plt.figure(figsize=(14, 8))
sns.violinplot(x='Continent', y='2023', data=df)
plt.title('Population Distribution by Continent (2023)')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.show()
```

```python
# Top 5 Countries with Largest and Smallest Population Changes
df['Population_Change_1950_2023'] = df['2023'] - df['1950']
top_5_largest_changes = df.nlargest(5, 'Population_Change_1950_2023')
top_5_smallest_changes = df.nsmallest(5, 'Population_Change_1950_2023')

plt.figure(figsize=(14, 8))
sns.barplot(x='Population_Change_1950_2023', y='Country', data=top_5_largest_changes)
plt.title('Top 5 Countries with Largest Population Changes (1950-2023)')
plt.xlabel('Population Change')
plt.ylabel('Country')
plt.show()

plt.figure(figsize=(14, 8))
sns.barplot(x='Population_Change_1950_2023', y='Country', data=top_5_smallest_changes)
plt.title('Top 5 Countries with Smallest Population Changes (1950-2023)')
plt.xlabel('Population Change')
plt.ylabel('Country')
plt.show()
```
